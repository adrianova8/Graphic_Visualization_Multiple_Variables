**MODAL DISTRIBUTION**


In addition to looking at whether a distribution is Skewed or Symmetric, histograms can be used to identify **Modes**.

A **mode** is the value with the most occurrences.

However, It is common to have **no** observations with the same value in a dataset, which makes, mode, useless for many real datasets.

A mode is represented by a **prominent peak** in the distribution. There is only one prominent peak in the histogram of **SIZE**.

Histogram that have one, two, or three prominent peaks are called respectively.

Any distribution with more than 2 prominent peaks is called **Multimodal**.

Notice that there was one prominent peak in the Unimodal distribution with a second less prominent peak that was since it only differs from its neighboring bins by a few observations ( Bills Histogram).

**VARIANCE**


The mean was introduced as a method to describe the center of a data set, but this is also important.

Below are the deviations for the 1st, 2nd, 3rd, and 244th observations in the total_bill variable. For computational convenience, the number of characters is listed in the thousands and rounded to the first decimal.

```python
# Observations of the variable total_bill

print(tips.total_bill.iloc[[0],])
print(tips.total_bill.iloc[[1],])
print(tips.total_bill.iloc[[2],])
print(tips.total_bill.iloc[[243],])
```

Here we are subtracting each observation of the variable total_bill for the mean , if then we square these deviation and divide by the total number observations, we achieve the **variance**.

\begin{align*}x_1\:−\:\bar{x} = 16.9 − 19.78 = -2.88\end{align*}
\begin{align*}x_2\:−\:\bar{x} =\hspace{2.5mm}10.3 − 19.78 = −9.48\end{align*}
\begin{align*}x_3\:−\:\bar{x} =\hspace{2.5mm}18.7 − 19.78 = -1.08\end{align*}
\begin{align*}.\end{align*}
\begin{align*}.\end{align*}
\begin{align*}.\end{align*}
\begin{align*}x_{244} − \bar{x} = 15.8 − 19.78 = -3.9\end{align*}

If we square these deviation and then take an average, the result is about equal to the sample variance, denoted by $s^2$:

$$\begin{align*}s^2\:=\:\frac{(-2.88)^2 + (−9.48)^2 + (-1.08)^2 + · · · + (-3.9)^2}{244 − 1} = 78.92\end{align*}$$

**STANDARD DEVIATION**

Standard Deviation is defined as the square root of the Variance : $%$$$\begin{align*}s = \sqrt{78.92} = 8.90\end{align*}$$$%$

```python
# The variance and standard deviation ('total_bill')

variance = round(tips['total_bill'].var(ddof=0),3)
standard_deviation = round(tips["total_bill"].std(),3)
print('The variance is: ', variance)
print('The standard deviation is', standard_deviation)
```

```python
# The variance and standard deviation ('tip')

variance = round(tips['tip'].var(ddof=0),3)
standard_deviation = round(tips["tip"].std(),3)
print('The variance is: ', variance)
print('The standard deviation is', standard_deviation)
```

## **Treat the DataSet**

```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 

warnings.filterwarnings('ignore')
```

```python
# Movies Dataset

movies_df = pd.read_csv("movies.dat", sep='::', header=None)
movies_df.head(3)
```

```python
#Drop the first column and rename the second and third colum

movies_df = movies_df.drop([0], axis=1)
movies_df.columns = ['movie', 'genre']
movies_df.head(3)
```

```python
#Separate the year of the film, remove the '()' and convert to a (int)

movies_df[['movie', 'year']] = movies_df['movie'].str.rsplit(pat='(', n=1, expand=True)
movies_df.year = movies_df.year.str[:-1].astype(int)
```

```python
# We have already created the genre list. Now, pandas can recognise the list

movies_df.genre = movies_df.genre.str.split('|').tolist()
movies_df.head()
```

```python
# We create a dictionary that explains how many movies have the same genre

genres_dicti = {} 
for i in movies_df["genre"]: 
    for j in i: 
        if j not in genres_dicti: 
            genres_dicti[j] = 1 # Create the new movie genre with one unit.
        else: 
            genres_dicti[j] += 1 # We add a movie genre unit to the dictionary if the genre repeats
            
genres_dicti
```

```python
#Create a matrix of booleans that said if the movie genre is T or F

genre = movies_df.genre
genre_index = genres_dicti

def bool_display_genre(boolean_genre, genre_name):
# Create empty dict
    bool_genres_dic = {}
    
    # Loop through all the tags
    for i, genre in enumerate(genre_name):
        
        # Apply boolean mask
        bool_genres_dic[genre] = boolean_genre.apply(lambda x: genre in x) # For the first column (Animation) it iterates all the rows with T or F if the genre matches the movie
            
    # Return the results as a dataframe
    return pd.DataFrame(bool_genres_dic)

genres_bool = bool_display_genre(genre, genre_index)

genres_bool.head(3)
```

```python
# Pearson Correlation with binary numbers

genres_corr = genres_bool.corr(method='pearson')
genres_corr.head(3)
```

```python
# Count how many times a genre is named along with all the other genres

genres_int = genres_bool.astype(int)
genres_int.head(3)

# Frequencies

genres_freq = np.dot(genres_int.T, genres_int)
genres_freq
```

```python
# Add labels and transform it back into a data frame.

genres_freq = pd.DataFrame(genres_freq, columns= genres_dicti.keys(), index = genres_dicti.keys())
genres_freq.head()
```

This Dataframe shows us the number of times that the same genre is repeated throughout our DataSet and also how many times two same genres coincide for the same movie.

For example, the Animation genre is repeated 105 times, that means there are 105 movies that have Animation genre. Below we see there are 84 movies featuring the genre Animation and Children's and so on.

```python
# Converting the boolean matrix to dichotomoic variable (1 or 0)

new_movie_df = genres_bool.astype(int)
new_movie_df.head(3)
```

```python
# Adding the movie and year columns to the new dataset

new_movie_df.insert(0, 'Movie', movies_df.movie)
new_movie_df.insert(1, 'Year', movies_df.year)

movies_df = new_movie_df #Change the name
```
